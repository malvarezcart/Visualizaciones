```python
import pandas as pd
import openpyxl
import os
import numpy as np

import matplotlib.pyplot as plt
```

```python
file = "sales data.csv"
```

```python
data = pd.read_csv(os.path.join(".", file))
data.head()
```

```python
data.to_excel(os.path.join(".", file.split(".")[0]+".xlsx"))
```

```python
data.info()
```

```python
# Hexagon Binning between Cost and Profit
x = data["Cost"]
y = data["Profit"]
xlim = x.min(), x.max()
ylim = y.min(), y.max()

plt.figure(figsize=(10,8))
plt.hexbin(x, y, gridsize=20, cmap='BuPu')
plt.xlim(xlim)
plt.ylim(ylim)
plt.xlabel("Product Cost")
plt.ylabel("Product Profit")
plt.title("Relation between Product Cost and Product Profit")
cb = plt.colorbar(label='Number of Products')
plt.savefig(os.path.join(".", "Product_Cost_VS_Profit_HexagonalBinning.png"))
plt.show()
```
